# 1ER EJERCICIO:
Obtén una palabra por medio del input y conviértela a puras minúsculas

```python
c=input()
c=c.lower()
print(c)
```

# 2DO ejercicio:
Posibles combinaciones de una cadena

```python
conmutables=set() #creo mi conjunto de conmutables
l=len(c)
for i in range (l-1):
    a=c[:i]+c[(i+1)]+c[i]+c[(i+2):]
    print(a)
    conmutables.add(a+'\n')
print(conmutables)
```

# 3ER EJERCICIO:
Enlista todas las posibles variaciones de la palabra que se obtengan al agregar una letra cualquiera del abecedario. La letra agregada puede ser inicial, final o entre dos de las letras existentes. Añade las variaciones al mismo conjunto ya creado.

```python
abc='abcdefghijklmnñopqrstuvwxyzáéíóú'
long=len(c)
half=int(long/2)
for x in abc:
    w=x+c
    print("inicio: ",w)
    conmutables.add(w+'\n')
    middle=c[:half]+x+c[half:]
    print("en medio: ",middle)
    conmutables.add(middle+'\n')
    f=c+x
    print("final: ", f)
    conmutables.add(f+'\n')
    
print(conmutables)
```

# 4to EJERCICIO:
Enlista todas las posibles variaciones de la palabra que se obtengan al eliminar una de las letras de la palabra. Añade las variaciones al mismo conjunto.

```python
print("Ahora si eliminamos una letra")
for i in range(l-1):
    delete= c[:i]+c[(i+1):]
    print(delete)
    conmutables.add(delete+'\n')
print(conmutables)
```

# 5to EJERCICIO:
Enlista todas las posibles variaciones de la palabra que se obtengan al intercambiar una de las letras de la palabra por
cualquier otra letra del abecedario y añade las variaciones al mismo conjunto.

```python
print("Ahora si intercambiamos una letra")
for x in abc:
    for i in range(l-1):
        exchange= c[:i]+x+c[(i+1):]
        print(exchange)
        conmutables.add(exchange+'\n')

print(conmutables)
    
```

# 6. Ejercicio: 
Abre el archivo “lemario.txt” y crea un conjunto nuevo con las palabras que vienen ahí.

```python
file=open('lemario.txt')
lemario=set()
for elementos in file:
    print(elementos)
    lemario.add(elementos)
    
```

# 7. Ejercicio: 
Si la palabra original viene en el lemario, muestra una lista conteniendo únicamente la palabra original.

```python
listas=[]
for p in lemario:
    if p == c:
        listas.append(p)
print(listas)
```
