# 1. 
a) Crea un mapa que contenga los datos del archivo paises.csv en el año 2013
Clave: El nombre del país
Valor: La expectativa de vida

```python
file=open('paises.csv')
lector=csv.reader(file)
lista1=[x for x in lector if x[2]=="2013"]
mapa2013=mapa()
for elemento in lista1:
    pais=(elemento[0])
    expectativa=float(elemento[1])
    mapa2013[pais]=expectativa
    print(pais, ":", mapa2013[pais])
```

f ) Calcula el promedio de expectativa de vida para todos los paises que empiezan
con C.

```python
file3=open('paises.csv')
lector3=csv.reader(file3)
C=[x for x in lector3]
mapaC=mapa()
cont=0
suma=0
promedio=0

for elemento in C:
    pais=(elemento[0])
    expectativa=float(elemento[1])

    if pais[0]=='C':
        cont+=1
        mapaC[pais]=expectativa
        #print(pais, ":", mapaC[pais])
        suma+=expectativa
        
promedio=suma/cont
print("El promedio es igual a: ",promedio)
```
